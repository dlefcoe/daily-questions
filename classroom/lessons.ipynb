```python
# In place of x and y while calling the function can we pass 
# two different numbers

# globals
PI = 3.14159

def area_of_circle(radius: float)->float:
    ''' return the area of a circle 
    
    args:
        radius (float): the radius of the circle

    retrun:
        the area of a circle
    '''
    
    # area = PI * r**2
    area = PI * pow(radius, 2)

    return area

# define radius
r1 = 10.0
r2 = 20

area_01 = area_of_circle(r1)
area_02 = area_of_circle(radius=r2)

print('the area of the circle is:', area_01)
print('the area of the circle is:', area_02)


# the total area
total_area = area_01 + area_02
print('the total area is: ', total_area)


total_area = area_of_circle(r1) + area_of_circle(r2)
print('function all in one place:', total_area)
print('function all in one place:', area_of_circle(r1) + area_of_circle(r2))


# lots of areas
list_of_areas = []
for i in range(10):
    area = area_of_circle(i)
    list_of_areas.append(area)

# # everything in one line of code
list_of_areas.clear()
for i in range(10): list_of_areas.append(area_of_circle(i))


print('the list of areas:', list_of_areas)
print('the sum of areas:', sum(list_of_areas))


# LEGB
# Local,  Enclosed,  Global,  Builtin



# can we also define return type of the function?   YES



```

```python
# Sum of n numbers squared

n1 = [1, 4, 6, 10, 20]
n2 = [0.02, 0.04 , 1.6 , 1.1]
n3 = [1, 3, 4, 5, 6, 9, 10, 19, 20]
n4 = [1, 3, 4, 5, 6, 9, 10, 19, 20, 20, 20, 20, 20]



def sum_of_squared_evens(numbers:list):
    ''' sum of the squares of only the even numbers
    
    args:
        numbers (list): list of numbers
    
    return:
        (float): sum of the even numbers squared
    '''
    if len(numbers) > 10:
        s = 'too much work for me'
        s = 0
        return s

    our_sum = 0
    for n in numbers:
        if n % 2 == 0:
            # this is even
            # our_sum = our_sum + n**2
            our_sum += n**2

    return our_sum

# outside function (back in the global scope)

result_a = sum_of_squared_evens(n1)
result_b = sum_of_squared_evens(n2)
result_c = sum_of_squared_evens(n3)
result_d = sum_of_squared_evens(n4)

print('the sum of the even squares:', result_a)
print('the sum of the even squares:', result_b)
print('the sum of the even squares:', result_c)
print('the sum of the even squares:', result_d)

# make a master list
all_lists = [n1, n2, n3, n4]

print('\nthe sum of the even squares:')
# total = 0
for x in all_lists:
    # total = total + sum_of_squared_evens(x)
    print(sum_of_squared_evens(x))

# print('the total is', total)

# When do we only write return (without a value)

```

```python
class Dog:
    ''' a class for dogs'''
    # shared with all the instances
    kind = 'canine'

    # specific to each instance
    def __init__(self, name:str) -> None:
        # the doc string that we see
        ''' something about dogs 
        
        name (str): the name of the dog
        
        '''
        self.name = name

# create an instance of the class
d = Dog('scruffy')
e = Dog('bob')

# shared attributes
print(d.kind)
print(e.kind)

# unique attributes
print(d.name)  
print(e.name)

# the doc string of the class
print(d.__doc__)

# this is the dict of the class
print(d.__dict__)


```

```python
class Dog:

    # shared with all the instances
    kind = 'canine'
    food_types = []    # <-- common to all
    
    # unique to each instance (constructor, initialise)
    def __init__(self, name, number_of_legs=4, 
                            its_first_trick='jump and spin') -> None:
        self.name = name
        self.tricks = [its_first_trick]
        self.number_of_legs = number_of_legs


    def do_a_trick(self, our_trick:str):
        ''' teach the dog do a trick 
        
        args:
            our_trick (str): a trick that the dog can learn

        appends the list of tricks.
        '''
        self.tricks.append(our_trick)
    

    def likes_food(self, my_food:str):
        ''' introduce a new food type 
        
        args:
            my_food (str): some food that the dog likes

        appends the food list.        
        '''
        self.food_types.append(my_food)

print('everythig about dogs is created here...')

```

```python
class Dog:

    # shared with all the instances
    kind = 'canine'
    food_types = []    # <-- common to all
    
    # unique to each instance
    def __init__(self, name, number_of_legs=4) -> None:
        self.name = name
        self.tricks = []
        self.number_of_legs = number_of_legs


    def do_a_trick(self, our_trick:str):
        ''' teach the dog do a trick '''
        self.tricks.append(our_trick)

    def likes_food(self, my_food):
        ''' introduce a new food type '''
        self.food_types.append(my_food)

    def __repr__(self) -> str:
        ''' user friendly output string '''
        s= '<class Dog>\n'
        s = s + f'name: {self.name}, '
        s = s + f'tricks: {self.tricks}'
        return s


# create some dog instances
d = Dog('scruffy')
e = Dog('bob')
f = Dog('fred', number_of_legs=3)


print(d)
```

```python
class Warehouse:
    # shared attributes
    purpose = 'storage'
    region = 'west'
    object_in_warehouse = ['table', 'chair', 'sofa']

    def __init__(self) -> None:
        self.object_in_warehouse=[]

    


# created instances. instantiate 
w1 = Warehouse
w2 = Warehouse

# so lets try this.... more desired behaviour
w1 = Warehouse()
w2 = Warehouse()


# if we change the shared attribute of one, then the other one changes too
w2.region = 'south'

# here is the proof
print(w1.__dict__)
print(w2.__dict__)

print(w1.region, w2.region)


w1.object_in_warehouse.append('bed')
print(w1.object_in_warehouse)
print(w2.object_in_warehouse)
print('---')
print(w1.__dict__)
print(w2.__dict__)


```
